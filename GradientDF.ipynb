```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y= (X-5)**2+3
print(X,Y)
```

```python
plt.plot(X,Y)
plt.ylabel("y")
plt.xlabel("X")
plt.show()
```

```python
#gradient descent
#initialize x with random value
x = 0
lr = 0.1

plt.plot(X,Y)
for i in range(50):
  grad = 2*(x-5)
  x = x-lr*grad
  y= (x-5)**2+3
  plt.scatter(x,y)
  print(x)


plt.show()


```
